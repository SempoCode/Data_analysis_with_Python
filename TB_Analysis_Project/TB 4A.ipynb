#**Cleaning the data**

```python
import pandas as pd
```

```python
#Loading data
file_path = "/content/4A.xlsx"

data = pd.read_excel(file_path)
data.head()
```

```python
data.info()
```

```python
#Removing features which are not going to be used for prediction
col_to_delete = ['INITIALS', 'ICD10 CODE', 'ICD10CODE', 'FINAL DIAGNOSIS', 'MARITAL STATUS']

data = data.drop(columns=col_to_delete)
data.info()
```

```python
data.shape
```

```python
#filtering to get samples with all kinds of TB as the working diagnosis
df_TB = data[data['WORKING DIAGNOSIS'].str.contains('TB', na=False)]
df_TB.shape
```

```python
df_TB.head()
```

```python
#function to compute the number of other diagnosis in a sample other than any kind of TB
def count_extra_elements(cell):
  elements = [element.strip() for element in cell.split(',') if element.strip()!='']
  return len(elements) - 1

#Computing the additional working diagnosis and creating a row for it
df_TB['NUMBER OF ADDITIONAL WORKING DIAGNOSIS'] = df_TB['WORKING DIAGNOSIS'].apply(count_extra_elements)
df_TB.sample(3)
```

```python
#converting the Date of admission column to DateTime format
df_TB['DATE OF ADMISSION'] = pd.to_datetime(df_TB['DATE OF ADMISSION'], errors='coerce')

#Extracting the month from the date and creating its column
df_TB['MONTH OF ADMISSION'] = df_TB['DATE OF ADMISSION'].dt.month

#Extracting the day of the week from the date and creating its column
df_TB['DAY OF ADMISSION'] = df_TB['DATE OF ADMISSION'].dt.strftime('%A')

#converting the date of discaharge column to DateTime format
df_TB['DATE OF DISCHARGE'] = pd.to_datetime(df_TB['DATE OF DISCHARGE'])

#Computing the days of stay and creating its column
df_TB['COMPUTED DAYS OF STAY'] = (df_TB['DATE OF DISCHARGE'] - df_TB['DATE OF ADMISSION']).dt.days

#filtering out samples with 28 or less days of stay
df_TB = df_TB[df_TB['COMPUTED DAYS OF STAY'] <= 28]

df_TB.sample(4)
```

##**Cleanig RELIGIONS**

```python
df_TB['RELIGION'].unique()
```

```python
#Grouping religions into Anglican, Moslem, Pentecoatal, Catholic, Others and unknown
religions = {'Protestant':'COU', 'C O U':'COU',
             'c o u': 'COU', 'C/U': 'COU', 'Islam':'ISLAM', 'MOSLEM':'ISLAM',
             'B/A':'PENTECOSTAL', 'SDA':'OTHERS','PRO':'OTHERS','PAG':'OTHERS',
             '?':'OTHERS', 'R/C':'CATHOLIC'
             }
df_TB['RELIGION'] = df_TB['RELIGION'].replace(religions)

#Filling the empty ones with unknown
df_TB['RELIGION'] = df_TB['RELIGION'].fillna('UNKNOWN')

df_TB['RELIGION'].unique()
```

```python
#Encoding the religions
df_TB['RELIGION'].replace({'ANGLICAN':1, 'MOSLEM':2, 'CATHOLIC':3, 'UNKNOWN':4,
                           'PENTECOSTAL':5, 'OTHERS':6}, inplace=True)
df_TB['RELIGION'].unique()
```

##**Cleaning AGE**

```python
df_TB['AGE'].unique()
```

```python
#Removing all samples with empty age
df_TB = df_TB.dropna(subset=['AGE'])

#removing all samples with non-numeric and unclear age value
df_TB = df_TB[~df_TB['AGE'].isin(['?', 'AD', '25+'])]

#Converting age datatype to integer
df_TB['AGE'] = df_TB['AGE'].astype(int)

df_TB['AGE'].unique()
```
